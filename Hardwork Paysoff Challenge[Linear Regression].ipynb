```python
x = x.values.reshape(-1)
y = y.values.reshape(-1)

# Verify the shapes of x and y
print(f"x shape: {x.shape}")
print(f"y shape: {y.shape}")
```

```python
# Define the hypothesis function for linear regression
# theta[0] is the intercept term and theta[1] is the slope term.
def hypothesis(x, theta):
    y_ = theta[0] + theta[1] * x
    return y_

# Compute the gradient of the cost function
# This function calculates the gradient with respect to the parameters theta.
def gradient(x, y, theta):
    m = x.shape[0]  # Number of training examples
    grad = np.zeros((2,))  # Initialize gradient
    for i in range(m):
        y_ = hypothesis(x[i], theta)  # Predicted value
        grad[0] += (y_ - y[i])  # Gradient for theta[0]
        grad[1] += (y_ - y[i]) * x[i]  # Gradient for theta[1]
    return grad / m  # Return the average gradient

# Perform gradient descent to minimize the cost function
# This function updates theta parameters iteratively to minimize the cost function.
def gradientdescent(x, y, max_steps=100, learning_rate=0.1):
    theta = np.zeros((2,))  # Initialize theta
    for step in range(max_steps):
        grad = gradient(x, y, theta)  # Compute gradient
        theta[0] -= learning_rate * grad[0]  # Update theta[0]
        theta[1] -= learning_rate * grad[1]  # Update theta[1]
    return theta

# Normalize x for better convergence
# Normalization helps in speeding up the convergence of gradient descent.
x = (x - x.mean()) / x.std()

# Train the model using gradient descent
theta = gradientdescent(x, y)

# Plot the data and the regression line
plt.style.use('grayscale')
plt.scatter(x, y, label='Data Points')  # Plot data points
plt.plot(x, hypothesis(x, theta), color='red', label='Regression Line')  # Plot regression line
plt.xlabel('X')
plt.ylabel('Y')
plt.title('Linear Regression using Gradient Descent')
plt.legend()
plt.show()

# Print the optimal values of theta found
print(f"Optimal theta: {theta}")
```
