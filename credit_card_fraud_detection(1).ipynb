Importing The Dependencies


```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split #yeah dataset ko 2 tukdo mae toodh dega ek training and ek testing ke liye
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score

```

```python
#loading the dataset to a pandas dataframe
credit_card_data=pd.read_csv('/content/drive/MyDrive/creditcard.csv')
```

```python
#first 5 rows of the dataset
credit_card_data.head()
```

```python
credit_card_data.tail()
```

```python
#dataset information
credit_card_data.info()
```

```python
#checking the number of missing values in each column
credit_card_data.isnull().sum()
```

```python
# distribution of legit transaction & fraudulent transaction
credit_card_data['Class'].value_counts()        #0 represents good transaction and 1 reprwsent bad this is unbalanced data kuki dono class mae bahhut difference hae
```

This Dataset Is Highly Unbalance




0----> Normal Transaction
1---> Fraudulant Transaction

```python
# separating the data for analysis
legit = credit_card_data[credit_card_data.Class == 0] #Class ka c capital hae
fraud = credit_card_data[credit_card_data.Class == 1]
```

```python
print(legit.shape)
print(fraud.shape)
```

```python
#statistical measures of the data
legit.Amount.describe()
```

```python
fraud.Amount.describe()
```

```python
#compare the values for both transactions
credit_card_data.groupby('Class').mean()
```

under-sampline

Build a sample dataset containing similar distributon of normal transactions and fraudulent transactions

Number of fraudulent transaction--. 492

```python
legit_sample=legit.sample(n=492)
```

concatenating  two dataframe

```python
new_dataset=pd.concat([legit_sample,fraud],axis=0)
```

```python
new_dataset.head()
```

```python
new_dataset.tail()
```

```python
new_dataset['Class'].value_counts()
```

```python
new_dataset.groupby('Class').mean()
```

**Spliting the Data Into Features & Targets**

```python
x = new_dataset.drop(columns='Class',axis=1)
y = new_dataset['Class']
```

```python
print(x)
```

```python
print(y)
```

**Split the data into training and testing data**

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,stratify=y,random_state=2)
```

```python
print(x.shape,x_train.shape,x_test.shape)
```

**Model Training**

```python
model=LogisticRegression(max_iter=10000)
```

```python
#training the logistic regression model with training data
model.fit(x_train,y_train)
```

model Evaluation

accuracy score

```python
x_train_prediction = model.predict(x_train)
training_data_accuracy = accuracy_score(x_train_prediction, y_train)
```

```python
print('accuracy on training data: ',training_data_accuracy)
```

```python
#accuracy on test data
x_test_prediction=model.predict(x_test)
test_data_accuracy=accuracy_score(x_test_prediction,y_test)
```

```python
print('accuracy score on test data:',test_data_accuracy)
```
